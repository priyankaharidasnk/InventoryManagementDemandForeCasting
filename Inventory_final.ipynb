```python
#Dropping columns which are not required for forecasting
df=df.drop(["Order","File_Type","SKU_number","SoldCount","ReleaseNumber"],axis=1)
```

```python
# Converting marketing type to categorical variable 
df['MarketingType'] = df['MarketingType'].astype('category')
df['MarketingType'] = df['MarketingType'].cat.codes
```

```python
#applying sclaing using StandardScaler

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
df_sc=pd.DataFrame(sc.fit_transform(df.drop(["SoldFlag","MarketingType","New_Release_Flag"],axis=1)),columns=df.drop(["SoldFlag","MarketingType","New_Release_Flag"],axis=1).columns)
df_sc[["SoldFlag","MarketingType","New_Release_Flag"]]=df[["SoldFlag","MarketingType","New_Release_Flag"]]
```

```python
df_sc
```

```python
#spliting to test and train data
from sklearn.model_selection import train_test_split

X=df_sc.drop(["SoldFlag"],axis=1)
y=df_sc.SoldFlag

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=32)
X_train.head()
```

```python
y_train.value_counts()
```

```python
#Applying SMOTE
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=32)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

print("Train data before SMOTE:\n" ,y_train.value_counts())
print("\nTrain data after SMOTE:\n" ,y_train_sm.value_counts())
```

```python
print("Total data count:" ,X_train_sm['StrengthFactor'].count())
```

```python
y_train_sm.value_counts()
```

```python
#Importing libraries for models and cross validation and evaluation matrics
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
import xgboost as xgb
```

```python
xgb_cl = xgb.XGBClassifier()
```

```python
y_predct_xgb=xgb_cl.fit(X_train_sm,y_train_sm).predict(X_test)
```

```python
print("******************XGBoost******************")
print (classification_report(y_test, y_predct_xgb,target_names = ['class 0', 'class 1']))
print ("Accuracy: {:.2f} %".format(accuracy_score(y_test, y_predct_xgb, normalize=True) * 100))
```

```python
sns.heatmap(confusion_matrix(y_test, y_predct_xgb), annot=True, fmt='.2f', cbar=False, linewidths=.5)
plt.xlabel("Predicted value")
plt.ylabel("Actual value")
plt.show()
```

```python
#Logistic regression model

lgr=LogisticRegression(solver='liblinear',penalty='l2' )

y_predct_lgr=lgr.fit(X_train_sm,y_train_sm).predict(X_test)

print("******************Logistic Regression******************")
print (classification_report(y_test, y_predct_lgr,target_names = ['class 0', 'class 1']))
print ("Accuracy: {:.2f} %".format(accuracy_score(y_test, y_predct_lgr, normalize=True) * 100))

sns.heatmap(confusion_matrix(y_test, y_predct_lgr), annot=True, fmt='.2f', cbar=False, linewidths=.5)
plt.xlabel("Predicted value")
plt.ylabel("Actual value")
plt.show()
```

```python
#Decision tree model

dtc=DecisionTreeClassifier(criterion='entropy', random_state=50)
y_predct_dtc=dtc.fit(X_train_sm,y_train_sm).predict(X_test)

print("******************Decision Tree******************")
print (classification_report(y_test, y_predct_dtc,target_names = ['class 0', 'class 1']))
print ("Accuracy: {:.2f} %".format(accuracy_score(y_test, y_predct_dtc, normalize=True) * 100))

sns.heatmap(confusion_matrix(y_test, y_predct_dtc), annot=True, fmt='.2f', cbar=False, linewidths=.5)
plt.xlabel("Predicted value")
plt.ylabel("Actual value")
plt.show()
```

```python
#Random forest

rfc=RandomForestClassifier(criterion='entropy', random_state=2)
y_predct_rfc=rfc.fit(X_train_sm,y_train_sm).predict(X_test)

print("******************Random Forest******************")
print (classification_report(y_test, y_predct_rfc,target_names = ['class 0', 'class 1']))
print ("Accuracy: {:.2f} %".format(accuracy_score(y_test, y_predct_rfc, normalize=True) * 100))

sns.heatmap(confusion_matrix(y_test, y_predct_rfc), annot=True, fmt='.2f', cbar=False, linewidths=.5)
plt.xlabel("Predicted value")
plt.ylabel("Actual value")
plt.show()
```

```python
#Multilayer perceptron (MLP)

mlpc = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000,activation = 'relu',random_state=3)
y_predct_mlpc = mlpc.fit(X_train_sm, y_train_sm.values.ravel()).predict(X_test)

print("******************MLP******************")
print (classification_report(y_test, y_predct_mlpc,target_names = ['class 0', 'class 1']))
print ("Accuracy: {:.2f} %".format(accuracy_score(y_test, y_predct_mlpc, normalize=True) * 100))

sns.heatmap(confusion_matrix(y_test, y_predct_mlpc), annot=True, fmt='.2f', cbar=False, linewidths=.5)
plt.xlabel("Predicted value")
plt.ylabel("Actual value")
plt.show()
```
